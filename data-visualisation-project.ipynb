```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import time

print("\n")
print("You have selected a Horizontal Bar Graph")
time.sleep(2)

# This function will execute if option 1 is selected 
def option_1():
    # Load the CSV file
    df = pd.read_csv('/kaggle/input/global-hunger-index-dataset/GHI.csv')
    
    # Extract the relevant column, ensuring it's numeric and handling missing values
    y = pd.to_numeric(df['2000'][15:25], errors='coerce')  # Convert to numeric, convert errors to NaN (Not a Number)
    y = y.dropna()  # Drop any NaN values (Not a Number) that result from coercion (implicit casting)
    
    # Check if y is empty after all NaNs are dropped 
    if y.empty:
        print("No valid numerical data available for plotting")
    else: 
        # Normalising the data values to a range suitable for the colormap
        norm = plt.Normalize(min(y),max(y))
        

    # Extract data for plotting
    x = df['Country'][15:25][y.index] # Only selects countries with valid y values
    
    
    # The 'viridis' colormap is applied to the normalised data
    sm = plt.cm.ScalarMappable(cmap = 'viridis',norm=norm)
    # Used to associate the colormap with the data being represented 
    sm.set_array([]) # The array is empty as the values are stored in the dataframe 
    
    # Adds a colorbar (like a key which uses colours to represent values)
    # sm (ScalarMappable) is used to convert numerical data into colours 
    # These colours are based on the cmap (colourmap) that has been chosen ('viridis')
    plt.colorbar(sm)
    
    # Create colors based on the colormap
    colors = sm.to_rgba(y)
    
    

    # Create horizontal bar plot
    plt.barh(x, y, color= colors)
    plt.title("Global Hunger of Different Countries in year 2000")
    plt.xlabel("Measure of Hunger")
    plt.ylabel("Countries")

    plt.show()

```

```python
import pandas as pd 
import matplotlib.pyplot as plt
import matplotlib as mpl
import time

print("\n")
print("You have selected a Scatter Plot")
time.sleep(2)

# This function will execute if option 3 is selected  
def option_3():

    # Load the CSV file
    df = pd.read_csv('/kaggle/input/global-hunger-index-dataset/GHI.csv')

    # Extract the relevant column, ensuring it's numeric and handling missing values
    y = pd.to_numeric(df['2000'][15:25], errors='coerce')  # Convert to numeric, convert errors to NaN (Not a Number)
    y = y.dropna()  # Drop any NaN values (Not a Number) that results from coercion (implicit casting)

    # Check if y is empty after all NaNs are dropped 
    if y.empty:
        print("No valid numerical data available for plotting")
    else: 
        # Normalising the data values to a range suitable for the colormap
        norm = plt.Normalize(min(y),max(y))

    # Extract data for plotting
    x = df['Country'][15:25][y.index] # Only selects countries with valid y values

    fig, ax = plt.subplots()
    scatter = ax.scatter(x, y, c=y, cmap='viridis')

    # The 'viridis' colormap is applied to the normalised data
    sm = plt.cm.ScalarMappable(cmap = 'viridis',norm=norm)
    # Used to associate the colormap with the data being represented 
    sm.set_array([]) # The array is empty as the values are stored in the dataframe

    # Adds a colorbar (like a key which uses colours to represent values)
    # sm (ScalarMappable) is used to convert numerical data into colours 
    # These colours are based on the cmap (colourmap) that has been chosen ('viridis')
    fig.colorbar(sm, ax=ax)
    # This line of code above places the colourbar on an axis

    # Create colors based on the colormap
    colors = sm.to_rgba(y)


    # Rotate x-axis labels, align right, and set rotation
    # This allows all country names to fit rather than being squished together
    plt.xticks(rotation=45, ha='right')

    # Create scatter plot
    plt.scatter(x, y, c = colors)
    plt.title("Global Hunger of Different Countries in year 2000")
    plt.xlabel("Countries")
    plt.ylabel("Measure of Hunger")


    plt.show()

```
